```python
df_2015['date'] = pd.to_datetime(df_2015['date'], errors='coerce') #converting to pandas date-time format
```

```python
df_2015 = df_2015.sort_values(by='date')
```

```python
df_2015.info()
```

```python
ranking_history = pd.DataFrame(columns=['player_name', 'rating', 'date']) #updated ranking after each match stored in ranking_history dataframe
```

```python

df_try=df_2015
```

```python
df_try.info()
np.size(df_try['player1_name'].unique())
```

```python
def get_k_factor(games_played, rating):
    if games_played < 30:
        return 40
    elif rating < 2400:
        return 20
    else:
        return 10
```

```python
#ranking_history[ranking_history['player_name']=='John Astley'].shape[0]
ranking_history.shape[0]
```

```python
for index, row in df_try.iterrows():
    player1 = row['player1_name']
    player2 = row['player2_name']
    match_date = row['date']

    # Filter past history
    filtered = ranking_history[ranking_history['player_name'] == player1]
    filtered2 = ranking_history[ranking_history['player_name'] == player2]

    # Initialize default ratings
    base_rating = 1000
    current_rating1 = base_rating
    current_rating2 = base_rating

    games1=1
    games2=1

    
    if not filtered.empty:
        max_date = filtered['date'].max() #latest date
        latest_entry = filtered[filtered['date'] == max_date].iloc[-1] #last entry of latest date
        current_rating1 = pd.to_numeric(latest_entry['rating'], errors='coerce')
        games1=filtered.shape[0]+1

    if not filtered2.empty:
        max_date = filtered2['date'].max()
        latest_entry2 = filtered2[filtered2['date'] == max_date].iloc[-1]
        current_rating2 = pd.to_numeric(latest_entry2['rating'], errors='coerce')
        games2=filtered2.shape[0]+1

    # Update ratings
    Ra=current_rating1
    Rb=current_rating2
    Ea = 1 / (1 + 10 ** ((Rb - Ra)/400))
    Eb=1-Ea
    Ka=get_k_factor(games1,Ra)
    Kb=get_k_factor(games2,Rb)
    
    new_rating1 = round(current_rating1 + Ka*(1-Ea))
    new_rating2 = round(current_rating2 + Kb*(0-Eb))

    # Append new ratings to ranking history
    new_row1 = pd.DataFrame({
        'player_name': [player1],
        'rating': [new_rating1],
        'date': [match_date]
    })
    new_row2 = pd.DataFrame({
        'player_name': [player2],
        'rating': [new_rating2],
        'date': [match_date]
    })

    ranking_history = pd.concat([ranking_history, new_row1, new_row2], ignore_index=True)
```

```python
ranking_history.info()
```

```python
cutoff_date = pd.to_datetime('2015-12-31')  # or any target date
df_filtered = ranking_history[ranking_history['date'] <= cutoff_date]
df_last = df_filtered.drop_duplicates(subset='player_name', keep='last')
df_last.sort_values(by='rating', ascending=False).head(100)
```
