In this notebook, we concatenate the matches dataset from Kaggle (1982-2020) and our scraped dataset (2020-2025).

```python
import pandas as pd
import numpy as np
```

```python
matches1 = pd.read_csv('Kaggle Data/matches (1982-2020) v2.csv')
matches2 = pd.read_csv('Scraped Data/matches (2020-2025) v1.csv')
```

```python
matches1.head()
```

```python
#
matches1 = matches1[['player1_name', 'player2_name', 'score1', 'score2', 'best_of', 'tournament_id', 'date', 'year']]
matches1.rename(columns={'player1_name': 'player1', 'player2_name': 'player2'}, inplace=True)
```

```python
matches2.head()
```

```python
matches2 = matches2[['Player1', 'Player2', 'Score1', 'Score2', 'Best Of', 'Tournament id', 'Match Date', 'Year']]
matches2.rename(columns={'Player1': 'player1',
                         'Player2': 'player2',
                         'Score1': 'score1',
                         'Score2': 'score2',
                         'Best Of': 'best_of',
                         'Tournament id': 'tournament_id',
                         'Match Date': 'date',
                         'Year': 'year'}, inplace=True)
```

```python
matches2.head()
```

```python
matches1.info()
```

```python
matches2.info()
```

```python
#Delete the common matches from matches1
matches2_ids = matches2['tournament_id'].unique()
print('Original length of matches1 is:', len(matches1))
for id in matches2_ids:
    matches1 = matches1[matches1['tournament_id'] != id]

print('After deleting common matches, the length of matches1 is:', len(matches1))

```

```python
matches = pd.concat([matches1, matches2], axis = 0, ignore_index = True)
```

```python
matches.to_csv('matches.csv', index=False)
```
