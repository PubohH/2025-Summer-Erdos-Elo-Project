In this notebook, we sort the 2020-2025 matches dataset by tournament start date.

```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('Scrape/matches (2020-2025).csv')
```

```python
matches.head()
```

```python
matches.info()
```

```python
#Drop the unnecessary column
matches.drop('Unnamed: 0', axis=1)
```

```python
#Create a column 'tournament_start_date'
def to_datetime_with_format(s):
    return pd.to_datetime(s[:10], format = '%Y-%m-%d')

matches['tournament_start_date'] = 0
for id in matches['Tournament id'].unique():
    row_index = matches[matches['Tournament id'] == id].index
    dates = matches.loc[row_index, 'Match Date']
    dates = dates.dropna().apply(to_datetime_with_format)
    start_date = dates.min()
    matches.loc[row_index, 'tournament_start_date'] = pd.to_datetime(start_date, format = '%Y-%m-%d')
    
```

```python
matches.head()
```

```python
#Sort the matches by tournament start date
matches = matches.sort_values(by = ['tournament_start_date', 'Unnamed: 0'])
```

```python
matches['index'] = range(len(matches))
matches.set_index('index', inplace=True)
matches.drop('Unnamed: 0', axis=1, inplace=True)
matches = matches.rename_axis(None)
```

```python
matches.head()
```

```python
matches.to_csv('matches (2020-2025) v1.csv')
```
